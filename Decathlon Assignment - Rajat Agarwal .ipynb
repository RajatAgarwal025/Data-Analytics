# Insight 1.

**#Calculating the RFM(recency, frequency and monetory) Analysis of the data:**

```python
# Create snapshot date
snapshot_date = df3['InvoiceDate'].max() + timedelta(days=1)
print(snapshot_date)
```

```python
# Grouping by Customer ID
data_process = df3.groupby(['Customer ID']).agg({
        'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
        'InvoiceDate1': 'nunique',
        'Net Amount': 'sum'})
data_process
```

```python
# Rename the columns 
data_process.rename(columns={'InvoiceDate': 'Recency',
                         'InvoiceDate1': 'Frequency',
                         'Net Amount': 'MonetaryValue'}, inplace=True)
```

```python
# Print top 5 rows and shape of dataframe
print(data_process.head())
print('{:,} rows; {:,} columns'
      .format(data_process.shape[0], data_process.shape[1]))
```

```python
# Scoring Recency (R)
# Define bins and labels for Recency
recency_bins = [-1, 30, 60, 90, 120, 180, float('inf')]  # 6 bins
recency_labels = [6, 5, 4, 3, 2, 1]  # 6 labels (1 for the lowest recency, 6 for the highest)

data_process['R_Score'] = pd.cut(data_process['Recency'], bins=recency_bins, labels=recency_labels, include_lowest=True)

# Scoring Frequency (F)
# Using quantiles to score Frequency
data_process['F_Score'] = pd.qcut(data_process['Frequency'].rank(method='first'), q=5, labels=[1, 2, 3, 4, 5], duplicates='drop')

# Scoring Monetary Value (M)
# Using quantiles to score Monetary Value
data_process['M_Score'] = pd.qcut(data_process['MonetaryValue'].rank(method='first'), q=5, labels=[1, 2, 3, 4, 5], duplicates='drop')

# Convert scores to integers
data_process['R_Score'] = data_process['R_Score'].astype(int)
data_process['F_Score'] = data_process['F_Score'].astype(int)
data_process['M_Score'] = data_process['M_Score'].astype(int)

# Calculate RFM Score
data_process['RFM_Score'] = data_process['R_Score'] + data_process['F_Score'] + data_process['M_Score']

# Function to categorize customers based on RFM score
def rfm_category(row):
    if row['RFM_Score'] <= 2:
        return 'New Customer'
    elif 3 <= row['RFM_Score'] <= 4:
        return 'At Risk'
    elif 5 <= row['RFM_Score'] <= 10:
        return 'Needs Activation'
    elif row['RFM_Score'] == 11:
        return 'Less Potential Customer'
    elif row['RFM_Score'] >= 12:
        return 'Loyal Customer'
    else:
        return 'Uncategorized'

# Apply the updated RFM categories function
data_process['Customer_Category'] = data_process.apply(rfm_category, axis=1)

# Print top rows to check the updated categories based on RFM_Score
print(data_process[['R_Score', 'F_Score', 'M_Score', 'RFM_Score', 'Customer_Category']].head())

# Display the updated DataFrame
data_process
```

# insight 2.

 **Customer segmentation based on monetary value to identify high-value, mid-tier, and low-value customers, 
 such that to redefine my AOV on the low value customers.**

```python
# Step 1: Calculate total spending for each customer
total_spending = data_process.groupby('Customer ID')['MonetaryValue'].sum().reset_index()
total_spending.rename(columns={'MonetaryValue': 'Total_Spending'}, inplace=True)

# Step 2: Merge total spending back to the original data_process DataFrame
data_process = data_process.merge(total_spending, on='Customer ID', how='left')

# Step 3: Define segmentation criteria
def spending_segment(row):
    if row['Total_Spending'] >= 300000:
        return 'High-Value Customer'
    elif row['Total_Spending'] >= 100000:
        return 'Mid-Tier Customer'
    else:
        return 'Low-Value Customer'

# Step 4: Assign segments based on total spending
data_process['Spending_Segment'] = data_process.apply(spending_segment, axis=1)

# Step 5: Print the top rows to check the segmentation
print(data_process[['Customer ID', 'Total_Spending', 'Spending_Segment']].head())

# Optionally, display the updated DataFrame
data_process
```

# Insight 3.
**#Top-Selling by quantity Products: Analyze the most frequently purchased products across both years.**

```python
import matplotlib.pyplot as plt 
from matplotlib.ticker import FuncFormatter


# Step 1: Convert 'InvoiceDate' to datetime if not already done
df3['InvoiceDate'] = pd.to_datetime(df3['InvoiceDate'])

# Step 2: Extract the year from 'InvoiceDate' (if not already done)
df3['Year'] = df3['InvoiceDate'].dt.year

# Step 3: Calculate total quantity for each product across all years
product_sales = df3.groupby('Description')['Customer ID'].count().reset_index(name='Total_Quantity')

# Step 4: Sort products by total quantity to find top-selling products
top_selling_products = product_sales.sort_values(by='Total_Quantity', ascending=False)

# Display top 10 selling products based on quantity
print("Top 10 Selling Products (by Quantity):")
print(top_selling_products.head(10))

# Step 5: Analyze sales trends for top-selling products based on quantity
# Filter for top products (e.g., top 5)
top_products = top_selling_products.head(5)['Description'].tolist()

# Create a DataFrame to analyze trends for top products
trend_data = df3[df3['Description'].isin(top_products)]

# Group by month and product to get sales trends
trend_data['Month'] = trend_data['InvoiceDate'].dt.to_period('M')  # Create a monthly period
monthly_trends = trend_data.groupby(['Month', 'Description'])['Customer ID'].count().unstack(fill_value=0)

# Step 6: Plot sales trends for top-selling products as a bar chart
plt.figure(figsize=(20, 10), dpi=100)  # Increased size and DPI for better visibility

# Create a bar chart
monthly_trends.plot(kind='bar', width=3.5, alpha=1.0)  # Adjust width and alpha for better visibility
plt.title('Sales Trends for Top-Selling Products (by Quantity)', fontsize=16)
plt.xlabel('Month', fontsize=16)
plt.ylabel('Total Quantity Sold', fontsize=16)

# Function to format y-axis labels
def thousands_formatter(x, _):
    return f'{int(x / 1000)}K' if x >= 1000 else str(int(x))

plt.gca().yaxis.set_major_formatter(FuncFormatter(thousands_formatter))

plt.xticks(rotation=45, fontsize=8)
plt.yticks(fontsize=10)
plt.legend(title='Product', bbox_to_anchor=(0.05, 1), loc='upper left', fontsize=12)
plt.grid(axis='y')  # Add grid only on y-axis for better readability
plt.tight_layout()  # Adjust layout to fit everything nicely
plt.show()
```

# Insight 4.
**#Total Customers across countries and Year wise.**

```python
import pandas as pd

# Step 1: Convert 'InvoiceDate' to datetime if not already done
df3['InvoiceDate'] = pd.to_datetime(df3['InvoiceDate'])

# Step 2: Extract the year from 'InvoiceDate'
df3['Year'] = df3['InvoiceDate'].dt.year

# Step 3: Group by year and country, counting unique customers
customer_count_by_year_country = df3.groupby(['Country', 'Year'])['Customer ID'].nunique().reset_index()

# Step 4: Pivot the data to have countries as index and years as columns
customer_count_pivot = customer_count_by_year_country.pivot(index='Country', columns='Year', values='Customer ID').fillna(0)

# Step 5: Add a total column for unique customers across all years
customer_count_pivot['Total_Customers'] = customer_count_pivot.sum(axis=1)

# Step 6: Sort the pivoted DataFrame by 'Total_Customers' in descending order
customer_count_pivot = customer_count_pivot.sort_values(by='Total_Customers', ascending=False)

# Display the updated pivoted DataFrame for confirmation
print(customer_count_pivot)
```

# Insight 5.
**#finding cancellations basis of country, so that specific actions on cancellations can be done country wise.**

```python
import pandas as pd

# Step 1: Extract the year from 'InvoiceDate'
df3['Year'] = df3['InvoiceDate'].dt.year

# Step 2: Identify canceled orders where Invoice starts with 'C'
canceled_orders = df3[df3['Invoice'].str.startswith('C', na=False)]

# Step 3: Group by year, user (Customer ID), and country, counting the number of cancellations
cancellation_counts = canceled_orders.groupby(['Year', 'Country']).size().reset_index(name='Cancellations')

# Step 4: Pivot the data to have Customer IDs as index and years as columns, grouped by Country
cancellation_pivot = cancellation_counts.pivot_table(index=['Country'], columns='Year', values='Cancellations', fill_value=0)

# Convert counts to integers
cancellation_pivot = cancellation_pivot.astype(int)

# Step 5: Add a Total Cancellations column
cancellation_pivot['Total_Cancellations'] = cancellation_pivot.sum(axis=1)

# Step 6: Sort the DataFrame based on Total Cancellations in descending order
cancellation_pivot = cancellation_pivot.sort_values(by='Total_Cancellations', ascending=False)

# Display the updated pivoted DataFrame for confirmation
print(cancellation_pivot)
```
